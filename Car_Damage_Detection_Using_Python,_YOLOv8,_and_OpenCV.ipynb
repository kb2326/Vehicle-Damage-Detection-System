```python
from PIL import Image

# Load the image
input_path = "/content/56.jpg"  # Replace with your image file path
output_path = "resized_image.jpg"  # Output file path

# Open the image
image = Image.open(input_path)

# Resize the image to 640x640
resized_image = image.resize((640, 640), Image.Resampling.LANCZOS)

# Save the resized image
resized_image.save(output_path)

print(f"Image resized and saved to {output_path}")
```

```python
import os
from PIL import Image

# Input and output folder paths
input_folder = "/content/input_folder"  # Replace with the folder containing your images
output_folder = "/content/output_folder"  # Replace with your desired output folder

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get the list of image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Resize the first 10 images in the folder
for i, image_file in enumerate(image_files[:40]):  # Limit to the first 10 images
    input_path = os.path.join(input_folder, image_file)
    output_path = os.path.join(output_folder, f"resized_{image_file}")

    # Open and resize the image
    with Image.open(input_path) as img:
        resized_img = img.resize((640, 640), Image.Resampling.LANCZOS)
        resized_img.save(output_path)

    print(f"Resized {image_file} and saved to {output_path}")

print("Resizing completed.")
```
